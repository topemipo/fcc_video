```python
# import neccessary libraries
import pandas as pd
from dotenv import load_dotenv
import os
```

```python
# load environment variables from .env file
load_dotenv()
filepath = os.getenv("filepath")

# Read the CSV file
df = pd.read_csv(filepath)
```

```python
# sort dataset by Document_ID
df['Query_ID_numeric']= df['Query_ID'].str.extract('(\d+)').astype(int)
df['Document_ID_numeric'] = df['Document_ID'].str.extract('(\d+)').astype(int)
df.sort_values(by='Document_ID_numeric', inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
# get the first 20 case docs and sort by query ID
first_20 = df[0:20].copy()
first_20.sort_values(by='Query_ID_numeric', inplace=True)
first_20.reset_index(drop=True, inplace=True)
first_20.head()

# # save the first 20 case docs with relevant queries to a new CSV file
# first_20.iloc[:,[0,1]].to_csv('first20cd_with_relevant_queries.csv', index=False)
```

```python
# get queries that are answered by more than one case docs
u = first_20.iloc[:,0].value_counts() 
print(len(u[u > 1].index.tolist())) # 5 queries are answered by more than one case docs

# get the queries that has more than one relevant casedocs
first_20[first_20.iloc[:,0].isin(u[u > 1].index.tolist())].sort_values(by='Query_ID_numeric')
```
